Importing data from Wikipedia List of Postal Codes of Canada

```python
website_url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
website_url.status_code
```

Pulling data from html file in website_url

```python
soup = BeautifulSoup(website_url.content, 'html.parser')
```

Finding the table that contains the postcode list using .find()

```python
myTable = soup.find('table')
```

Finding all data within td

```python
links = myTable.find_all('td')
links2 = list(links)
```

Putting the find_all result(td) in a list

```python
list_new=[]
for i in range(len(links2)):
    list_new.append(str(links2[i]))
    
```

Cleaning up the list

```python
list_new2=[]
for i in range(len(list_new)):
    
    list_new2.append(list_new[i][4:-5])
```

```python
list_new_upd=[]
char1 = 'e="'
char2 = '">'
for i in range(len(list_new2)):
    if list_new2[i][0] == '<':
        mystr = list_new2[i]
        mystr = mystr[mystr.find(char1)+3 : mystr.find(char2)]
        list_new_upd.append(mystr)
    elif list_new2[i][0] == 'N':
        list_new_upd.append(list_new2[i][0:12])
    else:    
        list_new_upd.append(list_new2[i])
list_new_upd
list_new2 = list_new_upd     
```

```python
import numpy as np
list_new2 = np.array(list_new2)
type(list_new2)
list_new3 = np.reshape(list_new2, (-1, 3))
list_new3
len(list_new3)
list_new3.shape[0]
```

Converting the cleaned list into a Pandas Dataframe

```python
import pandas as pd
list_new4 = pd.DataFrame(list_new3)
list_new4.columns = ["Postcode", "Borough", "Neighborhood"]      
#list_new4
```

Drop rows where Bourough is 'Not assigned'

```python
df_new = list_new4.drop(list_new4[(list_new4['Borough'] == 'Not assigned')].index)
#df_new
```

Resetting the dataframe in preparation to combine, group postcodes

```python
df_new2 = df_new.reset_index()
#df_new2
```

Grouped by Postcode then joined 'Neighborhood' column then reset the 'index'

```python
df_new3 = df_new2.groupby('Postcode').aggregate({'Borough': 'first', 
                             'Neighborhood': ', '.join }).reset_index()
df_new3
```

```python
df_new3.shape
```

Sort using the 'Postcode' column

```python
df_new4 = df_new3.sort_values(by=['Postcode'])
#df_new4
```

For Neighborhood columns that are 'Not assigned', it will have the same entry as the column 'Bourough'

```python
a = 0
for i in range(0,df_new4.shape[0]):
    if df_new4.iloc[i,2] == 'Not assigned':
        df_new4.iloc[i,2] = df_new4.iloc[i,1]
        a = a + 1
#df_new4
```

Get relevant part of JSON and transform it into a *pandas* dataframe

```python
# tranforming json file into a pandas dataframe library
from pandas.io.json import json_normalize

# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

Define information of interest and filter dataframe

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```
